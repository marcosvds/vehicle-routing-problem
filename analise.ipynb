```python
import matplotlib.pyplot as plt

x = [3, 4, 5, 6, 7, 8, 9, 10, 11]

# Dividindo os valores por 1000 para converter de ms para segundos
vetor_brute_force = [v / 1000 for v in [0, 0, 0, 1, 1, 2, 51500, 70034, 120007]]
vetor_hybrid_omp_mpi = [v / 1000 for v in [1003, 1003, 1003, 1003, 1003, 1003, 30574, 40206, 79499]]
vetor_local_search = [v / 1000 for v in [0, 0, 0, 0, 1, 1, 23863, 25090, 25032]]
vetor_mpi_distributed = [v / 1000 for v in [1002, 1003, 1002, 1003, 1003, 1003, 40319, 50261, 103293]]
vetor_openmp_threads = [v / 1000 for v in [0, 0, 1, 1, 1, 1, 29377, 35065, 68883]]

# Plotando os vetores com linhas contínuas
plt.plot(x, vetor_brute_force, label='Brute Force', marker='o', linestyle='-', color='b')
plt.plot(x, vetor_hybrid_omp_mpi, label='Hybrid OpenMP+MPI', marker='o', linestyle='-', color='g')
plt.plot(x, vetor_local_search, label='Local Search', marker='o', linestyle='-', color='r')
plt.plot(x, vetor_mpi_distributed, label='MPI Distributed', marker='o', linestyle='-', color='c')
plt.plot(x, vetor_openmp_threads, label='OpenMP Threads', marker='o', linestyle='-', color='m')

# Adicionando título e rótulos aos eixos
plt.title('Performance Comparison of Different Algorithms', fontsize=14, fontweight='bold')
plt.xlabel('Number of elements', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)

# Adicionando a legenda
plt.legend()

# Ajustando a grade
plt.grid(True)

# Mostrando o gráfico
plt.show()
```
