```python
import matplotlib.pyplot as plt
import numpy as np

label_x = [3, 4, 5, 6, 7, 8, 9, 10, 11]

# Dividindo os valores por 1000 para converter de ms para segundos
vetor_brute_force = [v / 1000 for v in [0, 0, 0, 1, 1, 2, 51500, 70034, 120007]]
vetor_hybrid_omp_mpi = [v / 1000 for v in [1003, 1003, 1003, 1003, 1003, 1003, 30574, 40206, 79499]]
vetor_local_search = [v / 1000 for v in [0, 0, 0, 0, 1, 1, 23863, 25090, 25032]]
vetor_mpi_distributed = [v / 1000 for v in [1002, 1003, 1002, 1003, 1003, 1003, 40319, 50261, 103293]]
vetor_openmp_threads = [v / 1000 for v in [0, 0, 1, 1, 1, 1, 29377, 35065, 68883]]

bar_width = 0.15
index = np.arange(len(label_x))

plt.bar(index, vetor_brute_force, bar_width, label='Brute Force')
plt.bar(index + bar_width, vetor_hybrid_omp_mpi, bar_width, label='Hybrid OpenMP+MPI')
plt.bar(index + 2 * bar_width, vetor_local_search, bar_width, label='Local Search')
plt.bar(index + 3 * bar_width, vetor_mpi_distributed, bar_width, label='MPI Distributed')
plt.bar(index + 4 * bar_width, vetor_openmp_threads, bar_width, label='OpenMP Threads')

plt.xlabel('Number of Nodes', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.title('Performance Comparison of Different Algorithms', fontsize=14, fontweight='bold')
plt.xticks(index + 2 * bar_width, label_x)
plt.legend()
plt.grid(True)

# Salvando a imagem na pasta imagens
plt.savefig('imagens/bar_chart_performance_comparison.png')

# Mostrando o grfico
plt.show()
```
